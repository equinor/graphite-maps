```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx

# Define grid dimensions
rows, cols = 8, 8
spacing = 1.0

# Generate staggered grid points
points = []
for i in range(rows):
    for j in range(cols):
        x = j * spacing + (i % 2) * (spacing / 2)
        y = i * (spacing * np.sqrt(3) / 2)
        points.append((x, y))

# Create graph and connect each node to adjacent nodes only
G = nx.Graph()
for i, (x, y) in enumerate(points):
    G.add_node(i, pos=(x, y))

# Connect only immediate neighbors
for i in range(rows):
    for j in range(cols):
        idx = i * cols + j
        if j < cols - 1:
            G.add_edge(idx, idx + 1)  # right neighbor
        if i < rows - 1:
            G.add_edge(idx, idx + cols)  # bottom neighbor
            if j < cols - 1:
                G.add_edge(idx, idx + cols + 1)  # bottom-right diagonal

# Plot the sparser triangular mesh
pos = nx.get_node_attributes(G, 'pos')
fig, ax = plt.subplots()
nx.draw(G, pos, with_labels=False, node_size=10, edge_color='gray', node_color='black', ax=ax)
plt.show()
```
